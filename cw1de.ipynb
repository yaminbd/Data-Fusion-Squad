```python
titles = []

base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
for i in range (1,51):
    scrape_url = base_url.format(i)

    result = requests.get(scrape_url)
    soup = bs4.BeautifulSoup(result.text, 'lxml')
    books = soup.select('.product_pod')

    for book in books:
        if len(book.select('.star-rating.Two')) != 0:
            title = book.select('a')[1]['title']
            titles.append(title)
```

```python
books = []

for i in range (1,4):
    url= f"https://books.toscrape.com/catalogue/page-{i}.html"
    response = requests.get(url)
    response.content
    soup = BeautifulSoup(response.text, 'html.parser')
    ol =soup.find('ol')
    articles = ol.find_all('article', class_='product_pod')
    books = []

    for article in articles:
      image = article.find('img')
      title = image.attrs['alt']
      star = article.find('p')
      star= star['class'][1]
      price = article.find('p', class_='price_color').text
      price = price[1:]
      books.append([title, price, star])
    
```

```python
df = pd.DataFrame(books,columns = ['Title', 'Price','Star Rating'])
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.amazon.co.uk/s?rh=n%3A404028&fs=true&ref=lp_404028_sar'
HEADERS = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36'}

response = requests.get(url, headers=HEADERS)
soup = BeautifulSoup(response.content, 'html.parser')

# The class names below are placeholders, you will need to inspect the actual page and use the correct ones.
book_containers = soup.find_all('div', class_='s-asin')  # Each book container

for book in book_containers:
    # Extract title
    title = book.find('span', class_='a-size-medium').get_text().strip() if book.find('span', class_='a-size-medium') else 'No Title Found'
    
    # Extract author, the class might need to be adjusted if it is not exact.
    author = book.find('a', class_="a-size-base a-link-normal s-underline-text s-underline-link-text s-link-style").get_text().strip() if book.find('a', class_="a-size-base a-link-normal s-underline-text s-underline-link-text s-link-style") else 'No Author Found'
    
    # Extract date, the class might need to be adjusted if it is not exact.
    date = book.find('span', class_='a-size-base a-color-secondary a-text-normal').get_text().strip() if book.find('span', class_='a-size-base a-color-secondary a-text-normal') else 'No Date Found'
    
    # Extract price, assuming the price is in a 'span' with class 'a-price'
    price_container = book.find('span', class_='a-price')  # Assuming this is the container for the price
    if price_container:
       price_span = price_container.find('span', class_='a-offscreen')  # Often used for the actual price
       price = price_span.get_text().strip() if price_span else 'No Price Found'
    else:
       price = 'No Price Found'

    rating_tag = book.find('i', {'class': 'a-icon-star-small'})
    rating_text = 'No Rating Found'  # Default text if rating is not found
    if rating_tag:
        rating_span = rating_tag.find('span', {'class': 'a-icon-alt'})
        if rating_span:
            rating_text = rating_span.get_text().strip()
    print(f"Title: {title}\nAuthor: {author}\nDate: {date}\nPrice: {price}\nRating: {rating_text}\n---")
```
